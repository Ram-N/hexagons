```python
fig, ax = plt.subplots(figsize=(20,20))
size=2
hg = HexGrid(12,15,size, flat=True)

COL2 = ['green', 'blue', 'orange', 'brown']

dist = size * np.random.uniform(0.2, 0.8)

for idx, h in enumerate(hg.hlist):    
    h.render(fill=None, lw=2)
    pts = h.get_points_center_rtheta(dist, 10)
    #The specific point is randomly chosen, out of the possible 6
    p = pts[np.random.randint(6)]
    h.connect_pt_to_vertices(p, color=COL2[3], lw=4)

plt.axis('equal')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 8)
    
plt.axis('off')

#save_file('EDGE_MID_PTS')
```

```python
fig, ax = plt.subplots(figsize=(20,20))
fc_bg = '#0C0C0C'
fig.patch.set_facecolor(fc_bg)



size=2
hg = HexGrid(12,15,size, flat=True)

COL2 = ['green', 'blue', 'orange', 'brown']
COL2 = RED_BROWNS_HIGH_SAT #+ YELLOWS_HIGH_SAT

for idx, h in enumerate(hg.hlist):    
    
    for l in range(10): #Let's do more interior points inside each Hex
        dist = size * np.random.uniform()
        theta = np.random.randint(0,90)
        pts = h.get_points_center_rtheta(dist, theta)
        #The specific point is randomly chosen, out of the possible 6
        p = pts[np.random.randint(6)]
        h.connect_pt_to_vertices(p, color=COL2[np.random.randint(len(COL2))], lw=2)

#Come with a strong border on top
for idx, h in enumerate(hg.hlist):            
    h.render(fill=None, lw=4, color='yellow', zorder=3)

plt.axis('equal')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 8)
    
plt.axis('off')

save_file(fig, 'INTERIOR_PTS_blkbg')
```
