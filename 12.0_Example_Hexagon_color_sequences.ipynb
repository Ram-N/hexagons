```python
#COL = BLUES_HIGH_SAT + RED_BROWNS_HIGH_SAT #+ PINKS_LOW_SAT
#COL = RED_BROWNS_HIGH_SAT
#COL = PINKS_LOW_SAT
#COL = YELLOWS_HIGH_SAT + GREENS_HIGH_SAT
#COL = BLACK_WHITES_LOW_SAT
COL = GREENS_HIGH_SAT + RED_BROWNS_HIGH_SAT


COL_SEQ = [RED_BROWNS, BLACK_WHITES_LOW_SAT,
           BLUES, GREENS, BROWNS]
#           BEIGES, BLUES, ORANGES, CYANS, YELLOWS]


COL_SEQ = [mcolor_names]

def get_next_color(fam, col):
    
    if np.random.random() < 0.11:
        fam+=1
        #col=0
        
    cfam = COL_SEQ[fam%len(COL_SEQ)]
    if np.random.random() < 0.15:
        col+=1
    
    color = COL_SEQ[0][col%len(COL_SEQ[0])]
    
    return(color, fam, col)

    return (cfam[col%len(cfam)], fam, col)
    
    
```

```python
fam, col = 0,0
for i in range(200):
    fc, fam, col = get_next_color(fam, col)
    print(fc, fam, col)
```
