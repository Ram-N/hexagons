# Hexagonal Reflections and Symmetry



This notebook illustrates the use of two methods.

- `h.get_lattice_points()`

This method serves two purposes. It will store the interior 'lattice' points of a hexagon in a dictionary called `lattice` which can then be used. It also returns the dictionary, for plotting purposes.

- `h.get_symmetrical_points((pt_name, x), axi)`

Given one point on the lattice, this method will return all 6 of its reflected siblings, or, if an `axi` is specified, will return that particular mirror reflected point. Note that this is MIRROR reflection, *not* rotational symmetry. Use axi = '6' for rotational symmetry.

```python
import logging
import matplotlib.pyplot as plt
import numpy as np

from hexagon_utilities import *
from colors import *
```

## 6-way symmetry

```python
fig, ax = plt.subplots(figsize=(10,10))

axi = '6'
pt_names = ['sa', 'sb', 'ea', 'eb', 'ab']

for y in range(5):
    pt_name = pt_names[y]
    plt.text(1, 2*y, pt_name, {'color': 'black', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
    for x in range(6):
        h = Hex(3+2*x, 2*y, size=1)
        h.render(fc='none', color='k') 
        h.get_lattice_points()
        symp = h.get_symmetrical_points((pt_name, x), axi)
        h.plot_points(symp,color='r')
        h.plot_points(h.lattice[pt_name][x], color='k')

        h.draw_axis(axi)

for t in range(6):
    plt.text(3+2*t, 10, str(t), {'color': 'blue', 'fontsize': 24, 'ha': 'center', 'va': 'center',
          'bbox': dict(boxstyle="round", fc="white", ec="black", pad=0.2)})
        
        
plt.title(f'Mirror Reflection of Lattice Points Across the {axi} Axis')        
plt.axis('scaled')
plt.xlim(0, 15)
plt.ylim(-2, 11);
```
